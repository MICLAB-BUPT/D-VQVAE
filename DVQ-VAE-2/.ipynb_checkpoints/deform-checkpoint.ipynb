```python


v = torch.tensor(meshB.vertices).unsqueeze(0).float()
f = torch.tensor(meshB.faces).unsqueeze(0)
v = v.detach() 
#f = f.detach() 
v = torch.autograd.Variable(v, requires_grad=True)
#f = torch.autograd.Variable(f, requires_grad=True)
mesh_obj = Meshes(verts=v , faces=f) 

map = torch.zeros(mesh_obj.verts_packed().shape[0])
print(map.size())
map[:100] = 1
map = map.bool()
#deform_verts = torch.full(mesh_obj.verts_packed().shape, 0.0, device='cpu', requires_grad=True)
deform_verts_ = torch.full([100,3], 0.0, device='cpu', requires_grad=True)
optimizer = torch.optim.SGD([deform_verts_], lr=0.01, momentum=0.8)

for i in range(0,100):  # non-learning based optimization steps
    optimizer.zero_grad()
    deform_verts = torch.full(mesh_obj.verts_packed().shape, 0.0, device='cpu', requires_grad=False)
    deform_verts[map] = deform_verts[map]+deform_verts_
    new_src_mesh = mesh_obj.offset_verts(deform_verts)
    loss_laplacian_smoothing = mesh_laplacian_smoothing(new_src_mesh)
    loss_mesh_edge = mesh_edge_loss(new_src_mesh)
    loss_mesh_normal_consistency = mesh_normal_consistency(new_src_mesh)
    
    loss = loss_mesh_edge*100#10*loss_laplacian_smoothing+loss_mesh_normal_consistency+
    loss.backward()
    optimizer.step()
    if i == 0 or i %10 == 0 :
        print("Object  iter {}, "
              "loss_laplacian_smoothing {:9.5f}, "
               "mesh_edge_loss {:9.5f}, "
             "loss_mesh_normal_consistency {:9.5f}, ".format(i,
                                           loss_laplacian_smoothing.item(),loss_mesh_edge.item()*100, loss_mesh_normal_consistency.item()))
```

```python
meshB.vertices=new_src_mesh.verts_packed().detach().numpy()
trimesh.Scene([meshB]).show()
```
